```python
df['Trend'] = 0
df = df.drop('Trend', axis=1)
```

```python
df['pct_change'] = df['close'].pct_change()

# set the threshold for a trend change (0.16% in this case)
threshold = 0.0016

# create a function to map the percentage change to a trend value
def get_trend(pct_change):
    if pct_change > threshold:
        return 1  # uptrend
    elif pct_change < -threshold:
        return -1  # downtrend
    else:
        return 0  # no change

# apply the get_trend function to the pct_change column to create the trend column
df['trend'] = df['pct_change'].apply(get_trend)

# drop the pct_change column if you don't need it anymore
df.drop('pct_change', axis=1, inplace=True)
```
