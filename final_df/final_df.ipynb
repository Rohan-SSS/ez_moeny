```python
import pandas as pd
import numpy as np

import technical_indicators as ti
```

```python
# df = pd.read_csv('..\\raw_data_30m\\currency_pairs\\EURUSD_M30.csv',  sep='\t')
df = pd.read_csv('../raw_data_30m/metal_cmx/XAUUSD_M30.csv', sep='\t')
```

```python
df = df.rename(columns={'Open':'open', 'High':'high', 'Low':'low', 'Close':'close', 'Volume':'volume'})
```

```python
df = df[-200000:]
df.reset_index(inplace=True)
df = df.drop(['index','Time'],axis=1)
```

```python
# Price change
ti.price_change(df)
# Getting EMA 
ti.ema_n(df, 100)
ti.ema_n(df, 200)
ti.ema_signal(df)
# VWAP
ti.vwap(df)
# S_RSI
ti.stochrsi(df)
# MACD
ti.macd(df)
# Trend
ti.get_trend(df)

df = df.dropna()
df.reset_index(inplace=True)
df = df.drop('index',axis=1)
```

```python
df = df.reindex(columns=['open', 'open_change', 'open_pct_change', 'high', 'high_change', 'high_pct_change', 'low', 'low_change',  'low_pct_change',
                        'close', 'close_change', 'close_pct_change', 'volume', 'volume_change', 'volume_pct_change', 
                        'ema_100', 'ema_200', 'vwap', 'STOCHk_14_3_3', 'STOCHd_14_3_3', 
                        'MACD_12_26_9',	'MACDs_12_26_9', 'MACDh_12_26_9', 
                        'ema_signal', 'vwap_signal', 'stochrsi_signal', 'macd_signal', 'trend'
    ])
```
